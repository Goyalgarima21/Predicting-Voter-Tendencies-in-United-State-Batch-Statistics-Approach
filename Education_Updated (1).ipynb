```python
zip_file_path = '/content/drive/MyDrive/ACSST5Y2020.S1501_2024-04-14T232508.zip'
df = gpd.read_file( 'zip://' + zip_file_path)
print(df.shape)
df.head()
```

```python
#Rename the columns using the values from the first row
df.columns = df.iloc[0]

# Drop the first row since it's now the column names
df = df.drop(0)

# Reset the index after dropping the first row
df = df.reset_index(drop=True)
df
```

```python
#print all the column
for column in df.columns:
    print(column)
```

```python
# Drop the Margin of Error
df.drop(columns=df.filter(like='Margin of Error').columns, inplace=True)


# Drop the last column has Nan values
df = df.iloc[:, :-1]
df
```

```python
# Check for missing values in the entire DataFrame
print(df.isnull().sum())
```

```python
#Replace Missing Values and Calculating Missing Value Proportions
df.replace('(X)', np.nan, inplace=True)
missing_values = df.isnull().sum() / len(df)
print(missing_values)
```

```python
#Identifying Columns with Missing Values
missing_columns = missing_values[missing_values >  0]
print(missing_columns)
```

```python
#Cleaning DataFrame by Dropping Columns with Missing Values
df_cleaned = df.dropna(axis=1)
df_cleaned
```

```python
def race_get_tract(string):
    return string.split(', ')[0][12:].upper()

def race_get_county(string):
    return string.split(', ')[1][:-7].upper()
```

```python
df_cleaned['County'] = df_cleaned['Geographic Area Name'].apply(race_get_county)
df_cleaned['Tract'] = df_cleaned['Geographic Area Name'].apply(race_get_tract)
df_cleaned[['Geographic Area Name', 'Tract', 'County']].head()
```
