```python
finaldata = pd.read_csv("/content/drive/MyDrive/Education_nc.csv")
finaldata
```

```python
bidenprop_county = pd.read_csv("/content/drive/MyDrive/bidenprop_county.csv")
#bidenprop_county.head()
bidenprop_county_ny = bidenprop_county[bidenprop_county['State'] == 'NORTH CAROLINA']
bidenprop_county_ny
```

```python
final_df1 = pd.merge(finaldata, bidenprop_county_ny, on='County')
final_df1
```

```python
columns_todrop = ['Tract','all_votes','State']
final_df2 = final_df1.drop(columns_todrop, axis =1)
final_df2.head()
```

```python
csv_file_path = 'finaldata_nc.csv'

# Use the to_csv method to save the DataFrame as a CSV file
final_df2.to_csv(csv_file_path, index=False)  # Set index=False to exclude the index column

print(f"DataFrame has been saved to {csv_file_path}")
```

```python
result_df = final_df2.groupby('Geographic Area Name')['Biden_proportion'].mean().reset_index()
result_df
```
